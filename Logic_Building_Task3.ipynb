```python
# Problem 1: Employee Performance Bonus Eligibility
```

```python
employees = {
"Ravi": 92,
"Anita": 88,
"Kiran": 92,
"Suresh": 85
}

totalEmp = [] # To store employess with highest salary

# Find highest score
maxScore = max(employees.values())

for name,score in employees.items():
    if score==maxScore:
        totalEmp.append(name)
        
#join combines all the values of the array separated by "," 

print(f"Top Performers Eligible for Bonus: {', '.join(totalEmp)} (Score: {maxScore})")
    
```

```python
# Problem 2: Search Query Keyword Analysis
```

```python
query = input("Enter your query: ")

query = query.lower()
words = query.split()

wordCount ={}

for word in words:
    wordCount[word] = wordCount.get(word,0) +1
    
repeatedWords = {}

for word,count in wordCount.items():
    if count>1:
        repeatedWords[word] = count

print(repeatedWords)
        
    
```

```python
# Problem 3: Sensor Data Validation
```

```python
# Sensor readings list
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

validReadings = {} # for storing the valid readings in the key-value pair

i=0 # to track the indexes of the readings 
for val in sensor_readings:
    if val %2==0: # check if the val is even
        validReadings[i] = val
    i+=1 #increases the index by 1 for every iteration
    
print(validReadings)
```

```python
# Problem 4: Email Domain Usage Analysis
```

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# to store the domain and its count
domainCount = {} 


for email in emails:
    domain = email.split("@")[1] # split the email at @ => [name@,domain] . here we need the second elements so index 1
    domainCount[domain] = domainCount.get(domain,0)+1 #Increase the domain count by 1

totalDomains = len(emails)

for domain , count in domainCount.items():
    percentage = (count / totalDomains) * 100 #Calcu;ate the percentage of each domain
    print(f"{domain} : {percentage}%")
```

```python
# Problem 5: Sales Spike Detection
```

```python
# Daily sales data
sales = [1200, 1500, 900, 2200, 1400, 3000]

salesSum = 0 # to store the total sum of the sales
for sale in sales:
    salesSum += sale

totalSales = len(sales) # get the total sales
avgSales = salesSum / totalSales # Store the avg sale value

i=1 # to track the days of the sale 
for sale in sales:
    if sale>=avgSales: # check if the sale is >= avg sales
        print(f"Day {i}: {sale}")
    i+=1
```

```python
# Problem 6: Duplicate User ID Detection
```

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

userIdCount = {} # to store the userId along with its count (key-value) pair 

for id in user_ids:
    userIdCount[id] = userIdCount.get(id,0) +1 #Increase the count of the userId by 1

for id,count in userIdCount.items():
    if count >1: # check if the userId had appread more than 1 
        print(f"{id} -> {count}")
```
