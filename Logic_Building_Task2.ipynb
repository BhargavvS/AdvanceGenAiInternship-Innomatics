```python
# Problem 1: Unique Words in a Sentence
```

```python
message = input("Enter a message: ")

#split the message into arrays of words using split operator
words = message.split(" ")

#To get all the unqiue values from the arrays we use "set" dataStructure.
#set datastructure stores ony the unique values

unique_words = set(words)

print("The number of unquie words are: ",len(unique_words))
print("The unique words are : " , unique_words)
```

```python
# Problem 2: Highest Salary from Employee Data
```

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

highest_emp = None # to store the emp name
highest_sal = 0 # to store the highest salary

for emp, sal in employees.items():
    if sal > highest_sal:
        highest_sal = sal
        highest_emp = emp
        
print(f"The highest salary employee is {highest_emp} with salary {highest_sal}")
```

```python
# Problem 3: Find Maximum and Minimum Values
```

```python
numbers = [45, 22, 89, 10, 66]

max_val = numbers[0]
min_val = numbers[0]

for number in numbers:
    if number > max_val:
        max_val = number
    if number < min_val:
        min_val = number

print("List:", numbers)
print("Max value:", max_val)
print("Min value:", min_val)

```

```python
# Problem 4: Count Products Above a Price Threshold
```

```python
prices = [450, 1200, 899, 1500, 300]

count = 0 # to store the product count value
for price in prices:
    if price > 1000:
        count += 1

print("Number of products above 1000: ", count)
```

```python
# Problem 5: Calculate Attendance Percentage
```

```python
attendance = ["P", "P", "A", "P", "P"]

# count method return the no of times the value present in the array
present_count = attendance.count("P")
total_days = len(attendance)

percentage = (present_count / total_days) * 100
print("Attendance Percentage:", percentage)
```

```python
# Problem 7: Count Character Frequency
```

```python
text = input("Enter your text: ")
char_freq = {} # map data structure which stores key-value pair 
# here key => character and value => frequency

for char in text:
    if char in char_freq:
        char_freq[char] += 1
    else:
        char_freq[char] = 1

print(char_freq)
```
