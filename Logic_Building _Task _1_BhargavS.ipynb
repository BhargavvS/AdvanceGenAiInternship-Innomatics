```python
marks = [45, 78, 90, 33, 60]

passed =0
failed =0

for mark in marks:
    if mark >= 50:
        passed+=1
    else:
        failed+=1

print("The number of students passed are: ",passed)
print("The number of students failed are: ",failed)
    
```

```python
messages = ["Hi", "Welcome to the platform", "OK"]

for message in messages:
    length = len(message)
    if length <10 :
        print(f"The length of {message} is {length}")
    else:
        print(f"The length of {message} is {length} --warning length greater than 10")
        
    
```
