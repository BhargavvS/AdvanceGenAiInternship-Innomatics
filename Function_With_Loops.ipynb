## Problem 1: Mobile Recharge Validation System

```python
def isValidRechargePlan(rechargePlan):
    if rechargePlan < 50:
        return False
    elif rechargePlan == 199:
        return True
    elif rechargePlan == 299:
        return True
    elif rechargePlan == 399:
        return True
    elif rechargePlan == 599:
        return True
    else:
        return False

while True:
    print("The available recharge plan : 199, 299, 399, 599")
    rechargePlan = int(input("Enter your recharge plan: "))

    isValid = isValidRechargePlan(rechargePlan)
    if isValid:
        print("Your recharge is successfull")
        break
    else:
        print("Invalid Recharge Plan \n")
        print("Please select a valid recharge plan \n")
```

##  Problem 2: Inventory Reorder Alert System

```python
def stockAlert(products):
    for product , quantity in products.items():
        if quantity < 15:
            print(f"{product} Stock Reorder Alert. Stock is < 15")
        else:
            print(f"{product} stock is ok")
products = {
    "Milk" : 25,
    "Curd" : 12,
    "Yogurt" : 18,
    "Eggs" : 90,
    "Panner" : 8
}

stockAlert(products)
```

# Problem 3: Student Result Processing System

```python
def resultsAnalysis(results):
    sum =0
    for result in results:
        sum += result
    totalStudents = len(results)
    avg = sum / totalStudents
    
    if avg >= 50:
        print("All students have passed the exams")
    else:
        print("All the students have failed the exam")
        
studentMarks = [60, 55, 40, 70]
resultsAnalysis(studentMarks)
```

# Problem 4: Cab Fare Estimator with Retry Option

```python
# here i am considering base fare 50 - upto 3kms , above 3kms 12 rupees per km

def calculateFair(kms , peakHour):
    totalFair = 0
    if kms <= 3:
        totalFair = 50
    else:
        fairAboveBase = (kms - 3) *12;
        totalFair = 50 + fairAboveBase
    
    if peakHour == "yes":
        totalFair = totalFair + (0.25 * totalFair)
        
    return totalFair

while True:
    kms = int(input("Enter the number of kms travelled: "))
    peakHours = input("Enter YES whether it was a peak hours else NO: ").lower()
    
    totalFair = calculateFair(kms, peakHours)
    print(f"Your total Fair is {totalFair} \n")
    
    nextFair = input("Enter YES if you want to calculate another fair else Enter NO: ").lower()
    
    if nextFair == "no":
        break
```

# Problem 5: Employee Attendance Eligibility Checker

```python
def employeeEligibility(attendences):
    totalPresent = 0
    for attendence in attendences:
        if attendence == "P":
            totalPresent +=1
    
    totalAttendence = len(attendences)
    
    attendencePercentage = (totalPresent / totalAttendence) * 100
    
    if attendencePercentage >= 75:
        print("Eligible")
    else:
        print("Not Eligible")

attendences = ["P", "P", "P", "P", "P", "A", "P"]
employeeEligibility(attendences)
```

# Problem 6: Password Strength Checker

```python
def checkPassword(password):
    digit = False
    special = False
    spc = "@#$"
    
    if len(password) < 8:
        return "Weak Password"
    
    for ch in password:
        if ch.isdigit():
            digit = True
        if ch in spc:
            special = True
    
    if digit and special:
        return "Strong Password"
    else:
        return "Weak Password"


password = input("Enter password: ")
print(checkPassword(password))
```
