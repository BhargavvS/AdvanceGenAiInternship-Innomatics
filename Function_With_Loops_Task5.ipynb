```python
def eng_chk(lik_lst):
    tot = 0
    for lik in lik_lst:
        tot += lik
    
    sts = "Viral Post" if tot >= 1000 else "Normal Engagement"
    print("Total Likes:", tot)
    print("Post Status:", sts)

arr = [200, 350, 500]
eng_chk(arr)
```

```python
def stk_chk(stk):
    sts = "Low Stock Alert" if stk < 10 else "Stock Sufficient"
    print("Medicine Stock:", stk)
    print("Status:", sts)

stockCount = int(input("Enter the stock count: "))
stk_chk(stockCount)
```

```python
def rnf_chk(rnf_lst, req):
    tot = 0
    for rnf in rnf_lst:
        tot += rnf
    
    avg = tot / len(rnf_lst)
    sts = "Adequate Rainfall" if avg >= req else "Inadequate Rainfall"
    print("Average Rainfall:", avg)
    print("Rainfall Status:", sts)

rnf_chk([70, 75, 71], 70)
```

```python
def dup_chk(usr_lst):
    sts = "Yes" if len(usr_lst) != len(set(usr_lst)) else "No"
    print("Duplicate Accounts Found:", sts)

arr = ["raj", "anu", "raj"]
dup_chk(arr)
```

```python
def age_chk(age):
    sts = "Eligible" if age >= 18 else "Not Eligible"
    print("Patient Age:", age)
    print("Eligibility Status:", sts)

age = int(input("Enter your age: "))
age_chk(age)
```

```python
def prm_chk(prc_lst):
    prm = []
    for prc in prc_lst:
        if prc > 2000:
            prm.append(prc)
    print("Premium Crops:", prm)
prices = [1500, 2500, 3200, 1800]
prm_chk(prices)
```

```python
def err_chk(err):
    if err == 0:
        sts = "Healthy"
    elif err <= 5:
        sts = "Minor Issues"
    else:
        sts = "Critical Issues"
    
    print("Error Count:", err)
    print("System Status:", sts)

err_chk(7)
```

```python
def txn_chk(amt):
    lim = 50000
    sts = "Approved" if amt <= lim else "Rejected"
    print("Transaction Amount:", amt)
    print("Transaction Status:", sts)

txn_chk(60000)
```

```python
def att_chk(att_lst):
    pre = 0
    for att in att_lst:
        pre += att
    
    per = (pre / len(att_lst)) * 100
    sts = "Eligible" if per >= 75 else "Not Eligible"
    print("Attendance Percentage:", per)
    print("Exam Eligibility:", sts)
attendences = [1,1,1,0,1]
att_chk(attendences)
```

```python
def bil_chk(unt):
    bil = 0
    
    if unt <= 100:
        bil = unt*3
    elif unt <= 200:
        bil = (100 *3) + ((unt -100) * 5)
    else:
        bil = (100 * 3) +(100 *5) + ((unt -200) *7) 
    if bil < 500:
        sts = "Low Usage"
    elif bil <= 1500:
        sts = "Moderate Usage"
    else:
        sts = "High Usage"
    
    return bil, sts

amt, sts = bil_chk(250)
print("Total Bill:", amt)
print("Usage Status:", sts)
```
